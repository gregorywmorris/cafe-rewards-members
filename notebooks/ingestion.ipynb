```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
# Connect to MySQL
engine = create_engine("mysql+mysqlconnector://root:mysql@localhost/caferewards")

# Query data
query = "SELECT * FROM customers"
customers = pd.read_sql(query, engine)

query = "SELECT * FROM events"
events = pd.read_sql(query, engine)

query = "SELECT * FROM offers"
offers = pd.read_sql(query, engine)

```

```python
# actual reward dollars tracked in events
offers = offers.drop(columns=['reward_dollars'])
```

```python
customers_events = pd.merge(customers, events, on='customer_id', how='inner')
```

```python
customers_events.head(1)
```

```python
df = pd.merge(customers_events, offers, on='offer_id', how='left')
```
